```python
import scipy.cluster.hierarchy as sch      # to build dendrogram and build the plotting
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
wine1 = pd.read_csv("wine.csv")
```

```python
wine1
```

```python
wine1.info()
```

```python
#customized normalization function
def norm_func(i):
    x1=(i-i.min())/(i.max()-i.min())
    return(x1)
```

```python
wine1['Alcohol'].min(),wine1['Alcohol'].max()
```

```python
(14.23-11.03)/(14.83-11.03)
```

```python
#Normalized dataframe (considering numerical part of data)
df_norm = norm_func(wine1.iloc[:,1:])
```

```python
df_norm
```

```python
#create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='single'))      #single linkage
```

```python
#create clusters
hc = AgglomerativeClustering(n_clusters = 2 , affinity = 'euclidean' , linkage = 'single')
```

```python
hc
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters = pd.DataFrame(y_hc,columns = ['Clusters'])
```

```python
y_hc
```

```python
Clusters
```

```python
wine1['h_clustered'] = Clusters
```

KMeans

```python
from sklearn.cluster import KMeans
```

```python
wine2 = pd.read_csv("wine.csv")
```

```python
wine2
```

```python
#Normalization / Standardization Function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_wine2_df = scaler.fit_transform(wine2.iloc[:,1:])
scaled_wine2_df
```

```python
wcss = []
for i in range(1,6):
    kmeans = KMeans(n_clusters = i , init = 'k-means++' , max_iter = 300 , random_state = 0)
    kmeans.fit(scaled_wine2_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,6),wcss)
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss
```

```python
kmeans = KMeans(n_clusters = 3 , random_state = 0)
predict = kmeans.fit_predict(scaled_wine2_df)
```

```python
predict
```

```python
clusters_new = KMeans(n_clusters = 2, random_state = 0)
new_predict = clusters_new.fit_predict(scaled_wine2_df)
```

```python
clusters_new.labels_
```
